```python
import pandas as pd
import numpy as np
import seaborn as sbs
import matplotlib.pyplot as plt
df=pd.read_csv("student-mat.csv",sep=',')
df.head()
```

```python
dfs=df[['schoolsup','famsup','paid','activities','nursery','higher','internet','romantic']]
dfs.head()
```

```python
dfs=dfs.replace('no', 0)
dfs=dfs.replace('yes',1)
dfs.head()
```

```python
n=np.array(dfs[['schoolsup','famsup','paid','activities','nursery','higher','internet','romantic']])
n=n.reshape(-1,5)
n.shape
```

```python
dfs
```

```python
m=np.array(dfs[['schoolsup','famsup','paid','activities','nursery','higher','internet','romantic']])
m=m.reshape(-1,5)
m.shape
```

```python
import numpy as np
from scipy.spatial import distance

# Step 1: Load your dataset as a contingency matrix (assuming you have it in `contingency_matrix`)

# Step 2: Select 5 objects (rows) from the contingency matrix
selected_objects = dfs[:5]

# Step 3: Calculate the distance matrix
dist_matrix = distance.cdist(selected_objects, selected_objects, metric='euclidean')
print(dist_matrix)


```

```python
from scipy.spatial import distance
dist_matrix = distance.cdist(n, m)
sbs.heatmap(dist_matrix)
plt.show()

```
